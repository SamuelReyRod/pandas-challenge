```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head(10)
```

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete_df['school_name'].unique())
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete_df['student_name'].count()
student_count
```

```python
# Calculate the total budget
#total_budget = 
#total_budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = student_data['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = student_data['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)].count()["student_name"]
student_count = school_data_complete_df['Student ID'].unique()
#print(passing_math_count)
#print(len(student_count))

passing_math_percentage = passing_math_count / len(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)].count()["student_name"]
student_count = school_data_complete_df['Student ID'].unique()
#print(passing_reading_count)
#print(len(student_count))

passing_reading_percentage = passing_reading_count / len(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete_df[
    (school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)
].count()["student_name"]
student_count = school_data_complete_df['Student ID'].unique()
overall_passing_rate = passing_math_reading_count /  len(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    'Total Schools': [school_count],
    'Total Students': [student_count],
    'Total Budget':[total_budget],
    'Average Math Score':[average_math_score],
    'Average Reading Score':[average_reading_score],
    '% Passing Math':[passing_math_percentage],
    '% Passing Reading':[passing_reading_percentage],
    '% Overall Passing': [overall_passing_rate]
})
district_summary
```

```python
#school budget
per_school_budget = school_data.set_index(["school_name"])["budget"]
per_school_budget
```

```python
#per student budget
per_student_budget = school_data.set_index(["school_name"])["budget"] / school_data.set_index(["school_name"])["size"]
per_student_budget
```

```python

reduced_df = school_data_complete_df[['school_name','Student ID','math_score']]
testing_df = reduced_df.groupby('school_name').agg({'Student ID':'count', 'math_score': lambda x: (x>=70).sum()})

math_score_df = (testing_df['math_score'] / testing_df['Student ID']) * 100
passing_math_score = math_score_df
passing_math_score
```

```python
#% passing reading (the percentage of students who passed reading)
reduced_df = school_data_complete_df[['school_name','Student ID','reading_score']]
testing_df = reduced_df.groupby('school_name').agg({'Student ID':'count', 'reading_score': lambda x: (x>=70).sum()})

reading_score_df = (testing_df['reading_score'] / testing_df['Student ID']) * 100
passing_reading_score = reading_score_df
passing_reading_score
```

```python
#Average math score 
reduced_data = school_data_complete_df[['school_name','math_score']]
average_math_score = reduced_data.groupby('school_name').mean().squeeze()
average_math_score
```

```python
#Average reading score
reduced_data = school_data_complete_df[['school_name','reading_score']]
average_reading_score = reduced_data.groupby('school_name').mean().squeeze()
average_reading_score
```

```python
# %overall passing (the percentage of students who passed math AND reading)--student cnt that passed math & read with >=70 / school student cnt
#math_reading = school_data_complete_df[['Student ID','school_name','reading_score','math_score']]
#math_reading
passing_df = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]
passing_df
grouped_df = passing_df.groupby("school_name").agg({'Student ID': 'count', 'math_score': 'count', 'reading_score': 'count'})
grouped_df
merged_df = pd.merge(grouped_df, school_data_complete_df.groupby("school_name").agg({'Student ID': 'count'}), on="school_name")
merged_df
merged_df["mean_passing_score"] = (merged_df["math_score"] + merged_df["reading_score"]) / 2
merged_df["mean_passing_score"]
merged_df["passing_rate_per_student"] = (merged_df["mean_passing_score"] / merged_df["Student ID_y"])*100
overall_passing = merged_df["passing_rate_per_student"]
overall_passing
```

```python
top_schools = merged_df = merged_df.sort_values('passing_rate_per_student', ascending=False)
top_schools.head()

```

```python
bottom_schools = merged_df = merged_df.sort_values('passing_rate_per_student', ascending=True)
bottom_schools.head()
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
summary_df = pd.DataFrame({
            #"School Name": school_name,
            "School Type": school_types,
            "TTL Students": school_size_students,
            "TTL School Budget":per_school_budget,
            "Per Student Budget": per_student_budget,
            "AVG Math Score":average_math_score,
            "AVG Reading Score":average_reading_score,
            "% Passing Math":passing_math_score,
            "% Passing Reading":passing_reading_score,
            "% Overall Passing":overall_passing,
            })

summary_df
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

summary_df['PSS Reclass'] = pd.cut(summary_df['Per Student Budget'], bins=spending_bins, labels=labels)
summary_reclass_df = summary_df
summary_reclass_df
```

```python
spending_math_scores = summary_reclass_df.groupby(["school_name","PSS Reclass"]).mean()["AVG Math Score"]
spending_reading_scores = summary_reclass_df.groupby(["school_name","PSS Reclass"]).mean()["AVG Reading Score"]
spending_passing_math = summary_reclass_df.groupby(["school_name","PSS Reclass"]).mean()["% Passing Math"]
spending_passing_reading = summary_reclass_df.groupby(["school_name","PSS Reclass"]).mean()["% Passing Reading"]
overall_passing_spending = summary_reclass_df.groupby(["school_name","PSS Reclass"]).mean()["% Overall Passing"]

summary_spending_df = pd.DataFrame({
    'AVG Math Score': spending_math_scores,
    'AVG Reading Score': spending_reading_scores,
    '% Passing Math': spending_passing_math,
    '% Passing Reading': spending_passing_reading,
    '% Overall Passing': overall_passing_spending
})

summary_spending_df.reset_index(inplace=True)
summary_spending_df = summary_spending_df.dropna(subset=['% Overall Passing'])
summary_spending_df
```
